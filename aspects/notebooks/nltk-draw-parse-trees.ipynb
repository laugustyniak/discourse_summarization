```python
def parse_and_draw(doc):
    discourse_parser = DiscourseParser()
    discourse_tree = discourse_parser.parse('tmp', parse_text=doc)
    print(discourse_tree)
    jupyter_draw_nltk_tree(discourse_tree, show_tree=True)
```
