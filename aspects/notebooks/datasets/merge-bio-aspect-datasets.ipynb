```python
def split_word_and_tag(text: str):
    words = text.split(' ')
    if len(words) > 2:
        print('-DOCSTART- -X- O O\n')
        for word_idx, word in enumerate(words):
            if not word_idx % 2:
                proper_word = word
            else:
                proper_tag = word
                print(proper_word + ' ' + proper_tag)
        print()
```
